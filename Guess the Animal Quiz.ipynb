```python
# creating function which check the answer
def Correction(guess, answer):
    global score
    still_guessing = True
    attempt = 0
    while still_guessing and attempt < 3:
        if guess.lower() == answer.lower():
            print("Correct Answer")
            score = score + 1 # updating score
            still_guessing = False
        else:
            if attempt < 2:
                guess = input("Sorry Wrong Answer, try again")
            attempt = attempt + 1 # updating attempts
    if attempt == 3:
        print("The Correct answer is ",answer )
    
score = 0
print("Guess the Animal")
guess1 = input("Name the India's National Animal?\n ")
Correction(guess1, "tiger")
guess2 = input("Name the National Bird?\n ")
Correction(guess2, "peacock")
guess3 = input("Name the King of the Jugle?\n ")
Correction(guess3, "lion")
print("Your Score is "+ str(score))
```
