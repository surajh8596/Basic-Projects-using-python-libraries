```python
import PyPDF2
```

#### Create File which we want to read

```python
pdf_reader=PyPDF2.PdfReader(open("pdf.pdf", 'rb'))
```

#### Create speach/Voice

```python
import pyttsx3
speaker=pyttsx3.init()
```

```python
for page_num in range(len(pdf_reader.pages)):
    text=pdf_reader.pages[page_num].extract_text()
    speaker.say(text)
    speaker.runAndWait()
speaker.stop()
```

#### Create mp3 file of pdf

```python
speaker.save_to_file(text,'audio.mp3')
speaker.runAndWait()
```
