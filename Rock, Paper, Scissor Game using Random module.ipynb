## Rock, Paper, Scissors game using random module

```python
import random
```

```python
## Logic of game:
    ## rock vs paper    --> paper wins
    ## rock vs scissor  --> rock wins
    ## paper vs scissor --> scissor wins
```

```python
l=["Rock", "Paper", "Scissor"]
while True:
    ccount=0
    ucount=0
    uc=int(input('''
Game Start..... 
1 Yes
2 No | Exit

         '''))
    if uc==1:
        for a in range(1,6):
            userInput=int(input('''
1 Rock
2 Paper
3 Scissor
                    '''))
            
            if userInput==1:
                uchoice="Rock"
            elif userInput==2:
                uchoice="Paper"
            elif userInput==3:
                uchoice="Scissor"
            
            cchoice=random.choice(l)
            if cchoice==uchoice:
                print("User_value", uchoice)
                print("Comp_Value", cchoice)
                print("Game Draw")
                ucount+=1
                ccount+=1
            elif (uchoice=="Rock" and cchoice=="Scissor")or(uchoice=="Paper" and cchoice=="Rock")or(uchoice=="Scissor" and cchoice=="Paper"):
                print("Comp_value", cchoice)
                print("User_Value", uchoice)
                print("You Won")
                ucount+=1
            else:
                print("Comp_value", cchoice)
                print("User_Value", uchoice)
                print("Computer Won")
                ccount+=1
        
        if ucount==ccount:
            print("User Score", ucount)
            print("Computer Score", ccount)
            print("Game Drawn....")
        elif ucount>ccount:
            print("User Score", ucount)
            print("Computer Score", ccount)
            print("You Won The Game....")
        else:
            print("User Score", ucount)
            print("Computer Score", ccount)
            print("Computer Won The Game....")
    
    else:
        break
```
